```python
import pandas as pd
pp = pd.read_csv('3425_data.csv')
```

```python
mini = pp
mini
```

```python
mini = mini.drop(['A4A5_agg','srcid', 'Mode'] , axis=1)
```

```python
ohen = pd.get_dummies(mini.astype(str), sparse = True)

```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpmax
from mlxtend.frequent_patterns import fpgrowth


frequent_itemsets = fpgrowth(ohen, min_support=0.5, use_colnames=True, verbose=0)
print("Done")
```

```python
frequent_itemsets1 = frequent_itemsets
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_itemsets1, metric="confidence", min_threshold=0.5)
```
