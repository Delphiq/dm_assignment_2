```python
import pandas as pd
df = pd.read_csv('2 ADA.ANUPoll33L.CSV.01464m.csv')
df_u = pd.read_csv('3425_data.csv')

# get the correlations for each attribute, remove duplicates, set to a 3 column frame, dort by highes correlation
p_full = df.corr(method="pearson").abs().unstack().sort_values(ascending=False).drop_duplicates().head(90).dropna() 
p_full = pd.DataFrame(p_full)
p_full[p_full.iloc[:, 0] > 0.5].head(50)
```

```python
s_full = df.corr(method="spearman").abs().unstack().sort_values(ascending=False).drop_duplicates().head(50).dropna()
s_full = pd.DataFrame(s_full)
s_full[s_full.iloc[:, 0] > 0.5].head(50)
```

```python
p_u = df_u.corr(method="pearson").abs().unstack().sort_values(ascending=False).drop_duplicates().head(50).dropna()
p_u = pd.DataFrame(p_u)
p_u[p_u.iloc[:, 0] > 0.5].head(50)
```

```python
s_u = df_u.corr(method="spearman").abs().unstack().sort_values(ascending=False).drop_duplicates().head(50).dropna()
s_u = pd.DataFrame(s_u)
s_u[s_u.iloc[:, 0] > 0.5].head(50)
```
